```python
# Import Libraries 
import pandas as pd
import seaborn as sns
import numpy as npy

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjusts the configuration of the plots we will create

# read in the data
df = pd.read_csv(r'C:\Users\student\Downloads\archive\movies.csv')
```

```python
# An overview of our data
pd.options.display.max_rows = 10
df.head()
```

```python
# Odering the data by gross value
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Drop rows with at least one null value
# using dropna() function
df.dropna(inplace=True)
```

```python
# Dropping duplicate rows
df.drop_duplicates()
```

```python
# Changing the datatype of bugdet and gross - to remove decimal places

df['budget'] = df['budget'].astype(npy.int64)
df['gross'] = df['gross'].astype(npy.int64)
```

```python
# Finding correlations
# Scatter plot with bugdet vs gross

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Film')
```

```python
# Plot the budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data = df, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
plt.show()
```

```python
# Let's start looking at the correlation
```

```python
df.corr(method='pearson', numeric_only=True)
```

```python
# High correlation between budget and gross
```

```python
correlation_matrix = df.corr(method='pearson', numeric_only=True)
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
# Company* vs Gross Earnings 

df_numeric = df

for col_name in df_numeric.columns:
    if (df_numeric[col_name].dtype == 'object'):
        df_numeric[col_name] = df_numeric[col_name].astype('category')
        df_numeric[col_name] = df_numeric[col_name].cat.codes

df_numeric.head()
```

```python
correlation_matrix = df_numeric.corr(method='pearson', numeric_only=True)
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for All Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
df_numeric.corr()
```

```python
correlation_matrix = df_numeric.corr()
corr_pairs = correlation_matrix.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs)>0.5]
high_corr
```
